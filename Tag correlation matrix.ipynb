```python
import string
import uuid
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd
import random
```

```python
tags = ["".join([random.choice(string.ascii_uppercase) for x in range(0,4)]) for z in range(0,10)]
```

```python
guids = [str(uuid.uuid4()) for x in range(0,100)]
```

```python
df = pd.DataFrame()
```

```python
df['id']=guids
```

```python
df['tags'] = [[random.choice(tags) for x in range(1,random.randint(2,10))] for zz in range(0, 100)]
```

```python
one_hot = MultiLabelBinarizer()
```

```python
one_hot.classes_
```

```python
df[one_hot.classes_] = one_hot.fit_transform(df['tags'])
```

```python
df.corr()
```
