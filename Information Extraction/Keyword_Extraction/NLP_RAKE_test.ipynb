## Try it on a dataframe

```python
df = pd.read_csv(r"D:\Work\Coding\Data\some-news.csv")
df = df.drop("Unnamed: 0", axis=1)
df = df.drop("Unnamed: 0.1", axis=1)
df['article'] = df['article'].astype(str)
```

```python
df.head()
```

```python
def getKeywords(i):
    keywords = rake.apply(i,text_for_stopwords=None)
    
    return keywords

def addKeywords(dataframe, columnName):

    extracted = []

    for i, row in dataframe.iterrows():
        extractedKeywords = getKeywords(row[columnName])
        extracted.append(extractedKeywords)

    dataframe["Keywords"] = extracted
    
    return dataframe
```

```python
df = addKeywords(df, "article")
```
